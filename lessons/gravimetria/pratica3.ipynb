## Terra Normal e anomalia ar-livre

Para calcular a anomalia Bouguer, precisamos calcular a anomalia ar-livre primeiro. Vamos utilizar os valores calculados na [prática 2](http://www.leouieda.com/geofisica1/lessons/gravimetria/pratica2.html).

```python
def gravidade_normal(lat):
    lat = (np.pi/180)*lat
    # WGS84 (Hofmann-Wellenhof and H. Moritz, 2006)
    a = 6378137 # metros
    b = 6356752.3142 # metros
    gamma_a = 9.7803253359 # m/s2
    gamma_b = 9.8321849378 # m/s2
    res = (a*gamma_a*np.cos(lat)**2 + b*gamma_b*np.sin(lat)**2)/np.sqrt((a*np.cos(lat))**2 + (b*np.sin(lat))**2)
    # Converter o resultado de m/s2 para mGal
    return utils.si2mgal(res)
```

```python
def gravidade_normal_em_p(lat, h):
    return gravidade_normal(lat) - 0.3086*h
```

```python
anom_arlivre = eigen.gravity_earth - gravidade_normal_em_p(eigen.lat, eigen.h_over_geoid)
```

```python
projetado.add_attribute('arlivre', anom_arlivre)
```

```python
mpl.figure(figsize=(18, 10))
projetado.contourf('arlivre', colorbar='horizontal', levels=80)
ax = mpl.gca()
ax.set_title("Anomalia ar-livre")
ax.coastlines()
```

## Anomalia Bouguer

Depois de remover a gravidade da Terra Normal calculada no ponto de medição (P), resta o efeito gravitacional da massa topográfica que ficou acima do elipsoide. Vamos aproximar esse efeito pela gravidade de uma **placa homogênea infinita** de espessura $t$ e densidade $\rho$:

$$ 
g_{BG} = 2 \pi G  \rho t 
$$

em que $G = 0.00000000006673\ m^3 kg^{-1} s^{-1}$ é a constante gravitacional.
Note que aqui utilizamos a **altitude da topografia $t$, não a altitude de medição $H$**.

Essa correção é válida somente para os continentes. 
Quando removemos a gravidade da Terra Normal ($\gamma_P$), 
assumimos que a Terra tem uma crosta com espessura média (~ 35 km) e densidade média ($\rho = 2.67\ g/cm^3$). 
Porém, nos oceanos há água ($\rho = 1\ g/cm^3$) e não crosta ($\rho = 2.67\ g/cm^3$).
Isso indica que removemos mais do que deveríamos.

Por isso, a correção deve ser diferente nos continentes e nos oceanos.
Podemos diferenciar os dois casos utilizando os valores de $t$. Se $t \geq 0$, estamos no continente, e se $t < 0$, estamos no oceano.

A **anomalia Bouguer** é anomalia ar-livre menos o efeito da massa topográfica ou da massa água:

$$
\Delta g_{BG} = g - \gamma + 0.3086H -
\begin{cases}
    2\pi G \rho_{crosta} t                      & \text{se } t \geq 0 \\
    2\pi G (\rho_{agua} - \rho_{crosta})(-t)    & \text{se } t < 0 \\
\end{cases}
$$

(Note que usamos $-t$ pois $t$ é negativo nos oceanos e não faria sentido utilizar uma espessura da placa infinita negativa.)

O arquivo `etopo1-0_5-mundo.gdf` contém o modelo topográfico [ETOPO1](http://www.ngdc.noaa.gov/mgg/global/global.html) calculado nos **mesmos pontos onde a gravidade foi medida**.
Este arquivo tem 3 colunas: longitude, latitude e altitude da topografia ($t$).

```python
etopo = Grid.load('../data/etopo1-0_5-mundo.gdf').set_projection(crs.Geodetic())
```

```python
mpl.figure(figsize=(18, 10))
etopo.reproject(crs.Mollweide(central_longitude=180)).pcolor('topography_shm', colorbar='horizontal', 
                                                             cmap=mpl.cm.terrain)
ax = mpl.gca()
ax.set_title(r"Topografia", fontsize=18)
ax.coastlines()
```

Agora podemos calcular a anomalia Bouguer correta para os continentes e oceanos.

```python
def plato_bouguer(t, rho_crosta, rho_agua):
    res = np.empty_like(t)
    res[t >= 0] = 2*np.pi*constants.G*rho_crosta*t[t >= 0]
    res[t < 0] = 2*np.pi*constants.G*(rho_agua - rho_crosta)*(-t[t < 0])
    return utils.si2mgal(res)
```

```python
anom_bouguer = anom_arlivre - plato_bouguer(etopo.topography_shm, rho_crosta=2670, rho_agua=1000)
```

```python
projetado.add_attribute('bouguer', anom_bouguer)
```

```python
mpl.figure(figsize=(18, 10))
projetado.contourf('bouguer', levels=80, colorbar='horizontal')
ax = mpl.gca()
ax.set_title(r"Anomalia Bouguer", fontsize=18)
ax.coastlines()
```
