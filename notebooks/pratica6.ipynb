## Tarefa 1: Gere um modelo de uma bacia

```python
area = [0, 100e3, 0, 10e3]
x = np.linspace(area[0], area[1], 100)
z = -np.ones_like(x)
```

```python
modelo_bacia = Moulder(area, x, z)
```

```python
modelo_bacia.run()
```

```python
modelo_bacia.save('modelo_bacia')
```

```python
modelo_bacia.plot()
```
