# Prática 4 - Anomalias de alguns lugares interessantes

```python
%matplotlib inline
from __future__ import division
from fatiando.gridder import Grid
from fatiando.vis import mpl
from fatiando import constants, utils
import fatiando
import numpy as np
from cartopy import crs
```

```python
print("Usando a versão do Fatiando: {}".format(fatiando.__version__))
```

```python
def gravidade_normal(lat):
    lat = (np.pi/180)*lat
    # WGS84 (Hofmann-Wellenhof and H. Moritz, 2006)
    a = 6378137 # metros
    b = 6356752.3142 # metros
    gamma_a = 9.7803253359 # m/s2
    gamma_b = 9.8321849378 # m/s2
    res = (a*gamma_a*np.cos(lat)**2 + b*gamma_b*np.sin(lat)**2)/np.sqrt((a*np.cos(lat))**2 + (b*np.sin(lat))**2)
    # Converter o resultado de m/s2 para mGal
    return utils.si2mgal(res)

def gravidade_normal_em_p(lat, h):
    return gravidade_normal(lat) - 0.3086*h

def plato_bouguer(t, rho_crosta, rho_agua):
    res = np.empty_like(t)
    res[t >= 0] = 2*np.pi*constants.G*rho_crosta*t[t >= 0]
    res[t < 0] = 2*np.pi*constants.G*(rho_agua - rho_crosta)*(-t[t < 0])
    return utils.si2mgal(res)
```

## Havaí

```python
havai = Grid.load('../data/eigen-6c3stat-havai.gdf').set_projection(crs.Geodetic())
havai.add_attribute('topo', Grid.load('../data/etopo1-havai.gdf').topography_shm)
```

```python
mpl.figure(figsize=(9, 7))
havai.reproject(crs.Mercator()).contourf('gravity_ell', levels=80)
ax = mpl.gca()
ax.coastlines()
ax.set_title('Gravidade do Havai a {:g} km de altitude'.format(0.001*havai.height[0]))
```

```python
mpl.figure(figsize=(9, 7))
havai.reproject(crs.Mercator()).contourf('topo', levels=80, cmap=mpl.cm.terrain)
ax = mpl.gca()
ax.coastlines()
ax.set_title('Topografia do Havai (m)')
```

```python
havai.add_attribute('arlivre', havai.gravity_ell - gravidade_normal_em_p(havai.lat, havai.height))
havai.add_attribute('bouguer', havai.arlivre - plato_bouguer(havai.topo, 2670, 1000))
```

```python
mpl.figure(figsize=(9, 7))
havai.reproject(crs.Mercator()).contourf('arlivre', levels=80)
ax = mpl.gca()
ax.coastlines()
ax.set_title('Anomalia ar-livre')
```

```python
mpl.figure(figsize=(9, 7))
havai.reproject(crs.Mercator()).contourf('bouguer', levels=80)
ax = mpl.gca()
ax.coastlines()
ax.set_title('Anomalia Bouguer')
```

```python
p1, p2 = [13, 196], [28, 207]
x, y = np.transpose([p1, p2])
```

```python
mpl.figure(figsize=(9, 7))
havai.reproject(crs.Mercator()).contourf('topo', levels=80, cmap=mpl.cm.terrain)
ax = mpl.gca()
ax.coastlines()
ax.plot(y, x, '-r', linewidth=2, transform=crs.Geodetic())
```

```python
perfil = havai.reproject(crs.Mercator()).profile(p1, p2, 200, projection=crs.Geodetic())
```

```python
fig, axes = mpl.subplots(2, 1, sharex=True, figsize=(8, 8))
ax1, ax2 = axes
ax2.plot(perfil.distance*0.001, perfil.topo, '-k')
ax2.set_ylabel('Topografia (m)')
ax2.set_xlabel('Distancia (km)')
ax2.grid(axis='y')
ax1.plot(perfil.distance*0.001, perfil.bouguer, '-r', label='Bouguer')
ax1.plot(perfil.distance*0.001, perfil.arlivre, '-b', label='Ar-livre')
ax1.legend(loc='center right', fancybox=True)
ax1.set_ylabel('Anomalia ar-livre e Bouguer (mGal)')
ax1.grid(axis='y')
ax1.set_xlim(perfil.distance.min()*0.001, perfil.distance.max()*0.001)
mpl.tight_layout(h_pad=0, pad=0)
```

## Hudson Bay

```python
hudson = Grid.load('../data/eigen-6c3stat-hudson.gdf').set_projection(crs.Geodetic())
hudson.add_attribute('topo', Grid.load('../data/etopo1-hudson.gdf').topography_shm)
```

```python
cent_lon = np.mean(hudson.area[2:])
proj = crs.NorthPolarStereo(central_longitude=cent_lon)
```

```python
mpl.figure(figsize=(9, 7))
hudson.reproject(proj).contourf('gravity_ell', levels=80)
ax = mpl.gca()
ax.coastlines()
```

```python
mpl.figure(figsize=(9, 7))
hudson.reproject(proj).contourf('topo', levels=80, cmap=mpl.cm.terrain, autoranges=False)
ax = mpl.gca()
ax.coastlines()
```

```python
hudson.add_attribute('arlivre', hudson.gravity_ell - gravidade_normal_em_p(hudson.lat, hudson.height))
hudson.add_attribute('bouguer', hudson.arlivre - plato_bouguer(hudson.topo, 2670, 1000))
```

```python
mpl.figure(figsize=(9, 7))
hudson.reproject(proj).contourf('arlivre', levels=80)
ax = mpl.gca()
ax.coastlines()
```

```python
mpl.figure(figsize=(9, 7))
hudson.reproject(proj).contourf('bouguer', levels=80)
ax = mpl.gca()
ax.coastlines()
```
