## Preparação

Antes de mais nada, vamos aproveitar as funções que utilizamos na [prática 3](http://nbviewer.ipython.org/github/leouieda/geofisica1/blob/master/notebooks/pratica3.ipynb) para calcular as anomalias ar-livre e Bouguer.

```python
%matplotlib inline
from __future__ import division
from IPython.display import HTML
from fatiando.gridder import Grid
from fatiando.vis import mpl
from fatiando import constants, utils
import fatiando
import numpy as np
from cartopy import crs
```

```python
print("Usando a versão do Fatiando: {}".format(fatiando.__version__))
```

```python
def gravidade_normal(lat):
    lat = (np.pi/180)*lat
    # WGS84 (Hofmann-Wellenhof and H. Moritz, 2006)
    a = 6378137 # metros
    b = 6356752.3142 # metros
    gamma_a = 9.7803253359 # m/s2
    gamma_b = 9.8321849378 # m/s2
    res = (a*gamma_a*np.cos(lat)**2 + b*gamma_b*np.sin(lat)**2)/np.sqrt((a*np.cos(lat))**2 + (b*np.sin(lat))**2)
    # Converter o resultado de m/s2 para mGal
    return utils.si2mgal(res)

def gravidade_normal_em_p(lat, h):
    return gravidade_normal(lat) - 0.3086*h

def plato_bouguer(t, rho_crosta, rho_agua):
    res = np.empty_like(t)
    res[t >= 0] = 2*np.pi*constants.G*rho_crosta*t[t >= 0]
    res[t < 0] = 2*np.pi*constants.G*(rho_agua - rho_crosta)*(-t[t < 0])
    return utils.si2mgal(res)
```

# Havaí

Nosso objetivo é mapear a topografia e as anomalias ar-livre e Bouguer para a região do Havaí:


```python
HTML('<iframe src="https://www.google.com/maps/embed?pb=!1m10!1m8!1m3!1d5916036.911229154!2d-157.7151201!3d20.5932929!3m2!1i1024!2i768!4f13.1!5e1!3m2!1sen!2sbr!4v1413401199263" width="600" height="450" frameborder="0" style="border:0"></iframe>')
```

Primeiro vamos carregar os dados de gravidade e topografia e juntar tudo em único conjunto chamado `havai`.

```python
havai = Grid.load('../data/eigen-6c3stat-havai.gdf', column_names=['lon', 'lat', 'height', 'gravity'])
havai.add_attribute('topo', Grid.load('../data/etopo1-havai.gdf').topography_shm)
havai.set_projection(crs.Geodetic())
```

Para facilitar a vida dos alunos (que não tem do que reclamar), vou salvar esse conjunto em um único arquivo [CSV](http://en.wikipedia.org/wiki/Comma-separated_values).
Como a bondade está transbordando, vou trocar os pontos por vírgulas para que o Excel em português não estrague o nosso dia.
O [site da prática](http://www.leouieda.com/geofisica1/lessons/gravimetria-praticas/pratica4.html) contém links para baixar esses arquivos.

```python
havai.dump_csv('../data/havai.csv')
! sed -i 's/\./\,/g' ../data/havai.csv
```

Agora posso calcular as anomalias ar-livre e Bouguer e adicioná-las no meu conjunto de dados (não sou tão bonzinho assim para dar as anomalias já calculadas).

```python
havai.add_attribute('arlivre', havai.gravity - gravidade_normal_em_p(havai.lat, havai.height))
havai.add_attribute('bouguer', havai.arlivre - plato_bouguer(havai.topo, 2670, 1000))
```

Vamos os mapas das anomalias e da topografia  usando a [projeção Mercator](http://en.wikipedia.org/wiki/Mercator_projection) e escalas de cor apropriadas.

```python
proj = havai.reproject(crs.Mercator())
fig, axes = mpl.subplots(1, 3, figsize=(14, 9), subplot_kw=dict(projection=proj.projection))
ax = axes[0]
proj.contourf('topo', levels=80, colorbar='horizontal', ax=ax, cmap=mpl.cm.BrBG_r)
ax.coastlines()
ax.set_title('Topografia')
ax = axes[1]
proj.contourf('arlivre', levels=80, colorbar='horizontal', ax=ax)
ax.coastlines()
ax.set_title('Anomalia ar-livre')
ax = axes[2]
proj.contourf('bouguer', levels=80, colorbar='horizontal', ax=ax)
ax.coastlines()
ax.set_title('Anomalia Bouguer')
mpl.tight_layout(pad=0, w_pad=0)
```

É muito útil olhar os dados alongo de um perfil que corte as principais feições dos mapas acima. Vamos estrair o perfil entre os pontos `p1 = [lat, lon]` e `p2` definidos abaixo (como mostrado no mapa).

```python
p1, p2 = [15, 197.5], [28, 207]

mpl.figure(figsize=(9, 7))
proj.contourf('topo', levels=80, cmap=mpl.cm.BrBG_r)
ax = mpl.gca()
ax.coastlines()
x, y = np.transpose([p1, p2])
ax.plot(y, x, 'o-k', linewidth=2, transform=crs.Geodetic())
```

```python
perfil = havai.reproject(crs.Mercator()).profile(p1, p2, 200, projection=crs.Geodetic())
```

Agora podemos fazer um gráfico com os 3 dados (topografia e anomalias) ao longo do perfil.

```python
fig, axes = mpl.subplots(2, 1, sharex=True, figsize=(14, 6))
ax1, ax2 = axes
d = perfil.distance*0.001
ax2.plot(d, perfil.topo, '-k')
ax2.hlines(0, d.min(), d.max(), linestyles='--')
ax2.set_ylabel('Topografia (m)')
ax2.set_xlabel('Distancia (km)')
ax2.grid(axis='y')

ax1.set_ylabel('Anomalia ar-livre e Bouguer (mGal)')
ax1.plot(d, perfil.bouguer, '-r', label='Bouguer')
ax1.plot(d, perfil.arlivre, '-b', label='Ar-livre')
ax1.hlines(0, d.min(), d.max(), linestyles='--')
ax1.legend(loc='center right', fancybox=True)
ax1.grid(axis='y')
ax1.set_xlim(d.min(), d.max())
mpl.tight_layout(h_pad=0, pad=0)
```

Perguntas para pensar:

* Por que a anomalia ar-livre é positiva nas ilhas?
* Por que a anomalia ar-livre tem seu mínimo logo ao lado das ilhas?
* Olhando para o perfil de topografia, por que após o baixo ao lado da ilha há um alto na topografia antes de estabilizar entre -5000 m e -6000 m?
* Como e por que as feições da anomalia ar-livre se correlacionam com a topografia?
* Por que a anomalia Bouguer é sempre positiva?
* Por que a anomalia Bouguer é menos positiva sobre as ilhas?

# Extra: Hudson Bay

Como um extra, vamos olhar como ficam as anomalias na região da Baía de Hudson, no Norte do Canadá:

```python
HTML('<iframe src="https://www.google.com/maps/embed?pb=!1m10!1m8!1m3!1d12110458.266253604!2d-85.99188428241526!3d61.3758016298575!3m2!1i1024!2i768!4f13.1!5e1!3m2!1sen!2sbr!4v1413404706398" width="600" height="450" frameborder="0" style="border:0"></iframe>')
```

```python
hudson = Grid.load('../data/eigen-6c3stat-hudson.gdf', column_names=['lon', 'lat', 'height', 'gravity'])
hudson.add_attribute('topo', Grid.load('../data/etopo1-hudson.gdf').topography_shm)
hudson.set_projection(crs.Geodetic())
```

```python
hudson.dump_csv('../data/hudson.csv')
!sed -i 's/\./\,/g' ../data/hudson.csv
```

```python
hudson.add_attribute('arlivre', hudson.gravity - gravidade_normal_em_p(hudson.lat, hudson.height))
hudson.add_attribute('bouguer', hudson.arlivre - plato_bouguer(hudson.topo, 2670, 1000))
```

Dessa vez, vamos usar uma [projeção polar esterográfica](http://en.wikipedia.org/wiki/Stereographic_projection) para fazer os mapas. 
A projeção Mercator distorce demais as regiões de altas latitudes.
Nesses casos, as [projeções cônicas](http://en.wikipedia.org/wiki/Map_projection#Conic) são melhores.

```python
proj = hudson.reproject(crs.NorthPolarStereo(central_longitude=np.mean(hudson.area[2:])))
fig, axes = mpl.subplots(1, 3, figsize=(14, 5), subplot_kw=dict(projection=proj.projection))
ax = axes[0]
proj.contourf('topo', levels=80, colorbar='horizontal', ax=ax, cmap=mpl.cm.BrBG_r)
ax.coastlines()
ax.set_title('Topografia')
ax = axes[1]
proj.contourf('arlivre', levels=80, colorbar='horizontal', ax=ax)
ax.coastlines()
ax.set_title('Anomalia ar-livre')
ax = axes[2]
proj.contourf('bouguer', levels=80, colorbar='horizontal', ax=ax)
ax.coastlines()
ax.set_title('Anomalia Bouguer')
mpl.tight_layout(pad=0, w_pad=0)
```

O perfil que vamos extrair vai do continente ao mar do Norte, cortando a baía.

```python
p1, p2 = [45, 260], [70, 300]

mpl.figure(figsize=(9, 6))
proj.contourf('topo', levels=80, cmap=mpl.cm.BrBG_r)
ax = mpl.gca()
ax.coastlines()
x, y = np.transpose([p1, p2])
ax.plot(y, x, 'o-k', linewidth=2, transform=crs.Geodetic())
```

```python
perfil = proj.profile(p1, p2, 200, projection=crs.Geodetic())
```

```python
fig, axes = mpl.subplots(2, 1, sharex=True, figsize=(14, 8))
ax1, ax2 = axes
d = perfil.distance*0.001
ax2.plot(d, perfil.topo, '-k')
ax2.hlines(0, d.min(), d.max(), linestyles='--')
ax2.set_ylabel('Topografia (m)')
ax2.set_xlabel('Distancia (km)')
ax2.grid(axis='y')

ax1.set_ylabel('Anomalia ar-livre e Bouguer (mGal)')
ax1.plot(d, perfil.bouguer, '-r', label='Bouguer')
ax1.plot(d, perfil.arlivre, '-b', label='Ar-livre')
ax1.hlines(0, d.min(), d.max(), linestyles='--')
ax1.legend(loc='upper left', fancybox=True)
ax1.grid(axis='y')
ax1.set_xlim(d.min(), d.max())
mpl.tight_layout(h_pad=0, pad=0)
```
