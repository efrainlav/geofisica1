## Terra Normal

Vamos calcular a gravidade da Terra Normal usando a fórmula:

$$
\gamma(\varphi) = \frac{a \gamma_a \cos^2 \varphi + b \gamma_b \sin^2 \varphi}{\sqrt{a^2 \cos^2 \varphi + b^2 \sin^2 \varphi}}
$$

Lembrando que $\varphi$ é a latitude. Vamos utilizar o elipsóde de referência WGS84. Valores numéricos de a, b, $\gamma_a$ e $\gamma_b$ foram obtidos do livro:

Hofmann-Wellenhof, B., and H. Moritz (2006), Physical Geodesy, 2nd, corr. ed. 2006 edition., Springer, Wien ; New York.

```python
def gravidade_normal(lat):
    lat = (np.pi/180)*lat
    # WGS84 (Hofmann-Wellenhof and H. Moritz, 2006)
    a = 6378137 # metros
    b = 6356752.3142 # metros
    gamma_a = 9.7803253359 # m/s2
    gamma_b = 9.8321849378 # m/s2
    res = (a*gamma_a*np.cos(lat)**2 + b*gamma_b*np.sin(lat)**2)/np.sqrt((a*np.cos(lat))**2 + (b*np.sin(lat))**2)
    # Converter o resultado de m/s2 para mGal
    return utils.si2mgal(res)
```

```python
gamma = gravidade_normal(lat)
```

```python
mpl.figure(figsize=(18, 10))
mpl.title("Gravidade Normal (calculada sobre elipsoide)")
mpl.pcolor(lon, lat, gamma, shape, basemap=bm, cmap=mpl.cm.Reds)
mpl.colorbar(orientation='horizontal', shrink=0.5, pad=0.01).set_label('mGal')
mpl.draw_coastlines(bm)
```

## Anomalia Ar-livre

A gravidade do elipsoide deveria ser calculada no ponto de observação (vamos chamar de ponto P), não sobre elipsoide. Vamos corrigir esse feito através da **correção ar-livre**:

$$
\gamma_P \approx \gamma - 0.3086H\ mGal
$$

em que $H$ é a altitude geométrica da medição. Esse valor é valido para $H$ em metros.

```python
def gravidade_normal_em_p(lat, h):
    return gravidade_normal(lat) - 0.3086*h
```

```python
gamma_p = gravidade_normal_em_p(lat, h)
```

```python
mpl.figure(figsize=(18, 10))
mpl.title("Gravidade Normal calculada no ponto P")
mpl.pcolor(lon, lat, gamma_p, shape, basemap=bm, cmap=mpl.cm.Reds)
mpl.colorbar(orientation='horizontal', shrink=0.5, pad=0.01).set_label('mGal')
mpl.draw_coastlines(bm)
```

Veja como $\gamma_P$ é praticamente igual a gravidade da Terra verdadeira (primeiro mapa). Isso mostra que as anomalias que vemos na gravidade da Terra são muito pequenas.

A **anomalia ar-livre** é a gravidade medida ($g$) menos a gravidade da Terra Normal calculada no ponto de medição ($\gamma_P$):

$$ 
\Delta g_{AL} = g - \gamma_P \approx g - \gamma + 0.3086H 
$$

```python
anom_arlivre = g - gamma_p
```

```python
amp = np.abs(anom_arlivre).max()
mpl.figure(figsize=(18, 10))
mpl.title(r"Anomalia ar-livre", fontsize=18)
mpl.pcolor(lon, lat, anom_arlivre, shape, basemap=bm, cmap=mpl.cm.RdBu_r, vmin=-amp, vmax=amp)
mpl.colorbar(orientation='horizontal', shrink=0.5, pad=0.01).set_label('mGal')
mpl.draw_coastlines(bm)
```
