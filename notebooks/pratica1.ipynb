# Exemplo de pontos aleatorios e linhas de voo

Dados gerado com uma função gaussiana 2D. Amostrado aleatoriamente, ao longo de linhas de voo e regularmente.

```python
%matplotlib inline
from fatiando.vis import mpl
from fatiando.gridder import Grid
from fatiando import utils
import fatiando
import numpy as np
```

```python
print("Usando a versão do Fatiando: {}".format(fatiando.__version__))
```

Vamos gerar alguns dados espalhados aleatoriamente e também espalhados em uma malha regular. Ambos são gerados pela mesma função, então deveriam ser bem parecidos.

```python
aletatorio = Grid.scatter((-5000, 5000, -5000, 5000), 100, z=-10, seed=6)
regular = Grid.regular((-5000, 5000, -5000, 5000), (100, 100), z=-10)
```

```python
aletatorio.add_attribute('dados', utils.gaussian2d(aletatorio.x, aletatorio.y, 10000, 1000, angle=45))
regular.add_attribute('dados', utils.gaussian2d(regular.x, regular.y, 10000, 1000, angle=45))
```

Vamos ver como ficam os dados irregulares quando usamos uma interpolação cúbica para fazer o mapa de contorno.

```python
aletatorio.interp((50, 50), algorithm='cubic').contourf('dados', levels=30)
aletatorio.points()
```

Qual é a diferença se usarmos uma interpolação linear?

```python
aletatorio.interp((50, 50), algorithm='linear').contourf('dados', levels=30)
aletatorio.points()
```

Para verificar qual dos dois está mais próximo, vamos fazer uma mapa do dados regularmente espaçado (que seria o mais próximo possível do vverdadeiro).

```python
regular.contourf('dados', levels=30)
```

Vou salvar os dados para poder abri-los no Geosoft.

```python
np.savetxt('../data/random-grid.txt', np.c_[aletatorio.x, aletatorio.y, aletatorio.z, aletatorio.dados], fmt='%g')
np.savetxt('../data/regular-grid.txt', np.c_[regular.x, regular.y, regular.z, regular.dados], fmt='%g')
```

# Gravidade do mundo

Os dados da gravidade do mundo foram baixados de http://icgem.gfz-potsdam.de/ICGEM/potato/Service.html usando o modelo EIGEN-6c3stat.

```python
from cartopy import crs
```

```python
eigen = Grid.load('../data/eigen-6c3stat-0_5-mundo.gdf').set_projection(crs.Geodetic())
```

Esses dados tem 4 colunas: longitude, latitude, altitude e gravidade. Vamos fazer um mapa da gravidade (em mGal) utilizando a [projeção Mollweid](http://en.wikipedia.org/wiki/Map_projection).

```python
mpl.figure(figsize=(18, 10))
plot = eigen.reproject(crs.Mollweide(central_longitude=180)).pcolor('gravity_earth', colorbar='horizontal')
ax = mpl.gca()
ax.set_title("Gravidade medida na superficie da Terra")
ax.coastlines()
```
