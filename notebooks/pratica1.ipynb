# Exemplo de pontos aleatorios e linhas de voo

Dados gerado com uma função gaussiana 2D. Amostrado aleatoriamente, ao longo de linhas de voo e regularmente.

```python
%matplotlib inline
from fatiando.vis import mpl
from fatiando import gridder, utils
import numpy as np
```

```python
area = (-5000, 5000, -5000, 5000)
shape = (100, 100)
```

```python
x, y, z = gridder.scatter(area, 100, z=-10, seed=6)
gx, gy, gz = gridder.regular(area, shape, z=-10)
```

```python
g = utils.gaussian2d(x, y, 10000, 1000, angle=45)
gg = utils.gaussian2d(gx, gy, 10000, 1000, angle=45)
```

```python
mpl.figure()
mpl.contourf(x, y, g, shape, 30, cmap=mpl.cm.Reds, interp=True)
mpl.plot(x, y, '.k')
```

```python
mpl.figure()
mpl.contourf(gx, gy, gg, shape, 30, cmap=mpl.cm.Reds)
```

```python
np.savetxt('../data/random-grid.txt', np.transpose([x, y, z, g]), fmt='%g')
np.savetxt('../data/regular-grid.txt', np.transpose([gx, gy, gz, gg]), fmt='%g')
```

# Gravidade do mundo

Baixados de http://icgem.gfz-potsdam.de/ICGEM/potato/Service.html

```python
lon, lat, h, g = np.loadtxt('../data/eigen-6c3stat-0_5-mundo.gdf', unpack=True, skiprows=34)
```

```python
shape = (361, 721)
```

```python
bm = mpl.basemap((0., 360, -90, 90), projection='robin')
```

```python
mpl.figure(figsize=(18, 10))
mpl.title("Gravidade medida na superficie da Terra")
mpl.pcolor(lon, lat, g, shape, basemap=bm, cmap=mpl.cm.Reds)
mpl.colorbar(orientation='horizontal', shrink=0.5, pad=0.01).set_label('mGal')
mpl.draw_coastlines(bm)
```
